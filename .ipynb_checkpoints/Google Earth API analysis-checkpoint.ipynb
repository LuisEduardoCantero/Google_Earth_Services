```python
import requests
import csv

# Su clave de API de Google Maps
api_key = 'AIzaSyDByTkYo_aCRRrh8w6tPsRIoZohcVcAZpc'

# La lista de códigos postales
codigos_postales = ['11000', '11100', '11200', '11300']

# Crear un archivo CSV para guardar la información
with open('ubicaciones.csv', 'w', newline='') as csvfile:
    fieldnames = ['codigo_postal', 'latitud', 'longitud']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    # Iterar sobre cada código postal
    for codigo_postal in codigos_postales:
        # Crear la URL de la solicitud
        url = 'https://maps.googleapis.com/maps/api/geocode/json?address={}&key={}'.format(codigo_postal, api_key)

        # Enviar la solicitud
        response = requests.get(url)

        # Procesar la respuesta
        data = response.json()

        # Obtener la latitud y longitud
        latitud = data['results'][0]['geometry']['location']['lat']
        longitud = data['results'][0]['geometry']['location']['lng']

        # Guardar la información en el archivo CSV
        writer.writerow({'codigo_postal': codigo_postal, 'latitud': latitud, 'longitud': longitud})
```

```python


path = f'/home/luiscantero/Hola/Python/Google_Earth_API/ventas/'
frames = []

for file in os.listdir(path):
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(path, file))
        df = df.loc[(df['Buyer Currency'] == 'MXN')]
        #df = df.dropna()
        #df['Buyer Postal Code'] = df['Buyer Postal Code'].apply(lambda x: x.zfill(5))
        frames.append(df[["Buyer Postal Code","Buyer Currency"]])

result = pd.concat(frames)
```

```python
def check_int(val):
    if type(val) != int:
        return False
    return True

if result['Buyer Postal Code'].apply(check_int).any() == False:
    print("La columna contiene valores que no son int.")
else:
    print("La columna solo contiene valores int.")
```
